```python
import pandas as pd
from ast import literal_eval

df = pd.read_csv('../data_news/weekly_data/test/behaviors_data.csv', converters={
                attribute: literal_eval
                for attribute in set([
                    'clicked_news', 'click_0_itemIDs'
                ])
            })
df.head()
```

```python
from random import shuffle

def shuffle_list(row):
    list_clicked_0 =  row['clicked'].split()
#     print(type(row['click_0_itemIDs']))
    list_can = [row['candidate_news']] + row['click_0_itemIDs'] 
#     print(row['clicked'], list_clicked_0)
    l =list(zip(list_can, list_clicked_0))
    shuffle(l)
    a, b = zip(*l)
#     print(list(a), list(b))
    row['candidate_news'] = list(a)
    row['clicked'] = ' '.join(list(b))
    
    return row
```

```python
# df.drop(['clicked'], axis='columns', inplace=True)
# df.rename(columns={'clicked_0':'clicked'}, inplace=True)
```

```python
from tqdm import tqdm
tqdm.pandas()

df = df.progress_apply(shuffle_list, axis=1)
df.head()
```

```python
df = df[df['user'].apply(lambda x: str(x).isdigit())].reset_index(drop=True)
```
